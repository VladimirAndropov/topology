# Контейнер `dict` (словарь)

**Dict** (словарь, ассоциативный массив, хеш-таблица) - *key-value* хранилище, т.е. способен хранить пары объектов "ключ-значение" и получать значение объекта по его ключу. В качестве ключа может использоваться любое *хэшируемое* значение.

image.png

### Зачем нужны словари?

Часто необходимо сопоставлять набор данных не с порядковыми числами, а с какими-либо другими данными.

Например, делаем программу для перевода с английского на русский. Нам неудобно хранить английские и русские слова в виде двух списков - при необходимости перевода слова его сначала надо найти в списке английских слов, запомнить его индекс,а затем посмотреть элемент с этим индексом во втором списке - **это неудобно**.

**Более того, это неэффективно**. Чтобы проверить, есть ли у нас перевод для определенного английского слова, нужно проверить все сохраненные английские слова!

Решить эти две проблемы и позволяет словарь (хеш-таблица).

Основные операции со словарем:
- добавить новую пару
- получить пару по ключу
- удалить пару по ключу

### Создание словарей

Создать словарь в Python можно либо с помощью фигурных скобок, либо с помощью функции `dict`:

Создадим пустой словарь:

```python
empty_dict_a = {}
empty_dict_b = dict()

# убедимся, что два варианта создания пустого кортежа эквивалентны:
print(empty_dict_a == empty_dict_b)
```

Итак, словарь — это способ хранить связи между различными объектами.  Давайте для примера создадим словарь, который будет хранить несколько простых русских слов и их английский перевод. Назовем такой словарь **rus_eng_dictionary**.

```python
rus_eng_dictionary = {'яблоко': 'apple', 'апельсин': 'orange' , 'бананы': 'bananas'}
```

Обычный бумажный русско-английский словарь хранит связи между русскими словами и английскими. Аналогично словарь в Python хранит пары элементов. Зная первый элемент пары, всегда можно получить второй по ключу. Так, узнать значение слова 'яблоко' при помощи словаря в Python можно следующим образом.

```python
rus_eng_dictionary['яблоко']
```

Мы получили перевод на английский — это слово 'apple'.

Итак, словарь в Python — структура данных, предназначенная для объединения связанной информации. Словари в языке Python представляют собой совокупность пар ключ–значение.

Существует несколько способов задания словаря в Python. Рассмотрим каждый из них подробно.

Первый способ создания словаря — с помощью фигурных скобок {}. Внутри фигурных скобок поочередно, через запятую прописываются пары ключ-значение.

```python
book = {'apples': 30, 'oranges': 20, 'bananas': 40}

print(book)

print(type(book))
```

Второй способ создания словаря — при помощи функции dict(). Внутри функции dict() прописывается ключ и присваивается его значение.

```python
book = dict(apples=30, oranges=20, bananas=40)

print(book)

print(type(book))
```

В словаре могут храниться данные разных типов. Существуют некоторые ограничения на ключи, и мы поговорим об этом позже, а пока что посмотрим на словарь, в котором хранятся разнотипные данные.

Создадим словари с парами:

```python
dict_example = {'ivan': 100, 'peter': 200, -10: 'new', (1, -10): True}  # Перечисляем пары через запятую
dict_example
```

```python
my_dict = {
    1: True,
    'apples': 30.3,
    ('oranges', '1 kilogram'): 20,
    'Ivan Ivanov': {
        'age': 23,
        'job': True,
        'position': 'programmer'
    }
}

print(my_dict)

print(type(my_dict))
```

Обратите внимание, что словарь можно записать не только в одну строчку, но и построчно, разделяя пары значений запятыми.

В приведенном выше примере мы использовали типы int, boolean, str, tuple, float и даже словарь (dict) внутри словаря!

Зачастую нам не нужно наполнять наш словарь в момент его создания. Например, мы хотим просто создать словарь, а уже затем, используя алгоритм, наполнить его. Тогда нам пригодятся знания о том, как создать пустой словарь.

Самым удобным способом создать словарь из двух списков (список ключей и список значений) является использование функции `zip`:

```python
keys = ['ivan', 'peter', -10, (1, -10)]
values = [100, 200, 'new', True]

dict_example = dict(zip(keys, values))
dict_example
```

```python
a = [1, 2, 3, 4]
b = (1.5, 2.5, 3.5, 4.5)
c = "АБВГД"
res = list(zip(a, b, c))  # дополнительно преобразуем в список, т.к. zip возвращает особый тип данных
res
```

```python
res[2][0]
```

Если хотим создать словарь с заданным набором ключей и одинаковыми (начальными) значениями, удобно использовать метод `fromkeys`:

```python
keys_list = ["Январь", "Февраль", "Март", "Апрель", "Май", "Июнь", "Июль", "Август", "Сентябрь", "Октябрь", "Ноябрь", "Декабрь"]
month_dict = dict.fromkeys(keys_list)  # необычное написание, вызываем метод от "функции" dict, а не от словаря
month_dict
```

Если хотим другое начальное значение, то можем указать его вторым аргументом:

```python
month_dict = dict.fromkeys(keys_list, [])
month_dict
```

### Методы словарей

**Что будет, если запросим значени ключа, которого нет в словаре?**

Получим ошибку:

```python
dict_example['ivan_the_greatest']
```

Аналогичную ошибку получим при попытке удалить пару по отсутствующему ключу.

Такая ситуация может возникать достаточно часто, а ошибки в программе - вещь неприятная. Можем ли мы как-нибудь *безопасно* получить значение по ключу, если он есть?

Можем, для этого есть 2 метода:

Рассмотрим некоторые полезные методы словарей, полный список методов с описанием - [ссылка](https://python-reference.readthedocs.io/en/latest/docs/dict/).

Для безопасного получения значения из словаря используем методы `get` и `setdefault`:

```python
dict_example = {'ivan': 100, 'peter': 200, -10: 'new', (1, -10): True, 'oleg': 42}
dict_example
```

1. метод `get` - если ключ есть, то возвращает соответствующее значение, если нет, то возвращает `None`:

```python
key = 'peter'
dict_example.get(key)
```

```python
print(dict_example.get('alex'))
```

Если хотим, чтобы при отсутствии значения возвращалось что-то другое, можем указать это как второй аргумент:

```python
dict_example.get('alex', "not_found")
```

```python
dict_example
```

2. метод `setdefault` - если ключ есть, то возвращает соответствующее значение, если нет, то возвращает `None` **И** добавляет это значение для запрашиваемого ключа в словарь:

```python
key = 'peter'
dict_example.setdefault(key)
```

```python
dict_example.setdefault('alex')
```

```python
dict_example
```

Аналогично, можем указать значение "по умолчанию" вторым аргументом метода:

```python
dict_example.setdefault('dmitry', "not_present")
```

```python
dict_example
```

Можно узнать количество записей в словаре с помощью метода len()

```python
my_dict = {1: 'mango', 2: 'pawpaw'}
print(len(my_dict))

my_dict = dict()
print(len(my_dict))
```

Функция list() не только создает новый список, но также может преобразовывать словарь в список. Однако получится список, составленный только из ключей.

```python
my_dict = {1: 'mango', 2: 'pawpaw', 3: 'three'}

my_keys = list(my_dict)

for key in my_keys:

    print(my_dict[key])
```

Словари можно сравнивать между собой. Однако сравнение допустимо только на равенство == и неравенство !=. Равные словари имеют одинаковое количество пар и при этом в соответствующей паре совпадают ключ–значение. Сравнение при помощи операторов <, >, >=, <= приведет к ошибке TypeError.

```python
a = {1: 'mango', 2: 'pawpaw', 3: 'three'}

b = {2: 'pawpaw', 1: 'mango', 3: 'three'}

print(a == b)

c = {1: 'mango', 3: 'three'}

print(c == a)

print(c != b)

print(c < a)
```

Метод .get() позволяет получить значение ключа. Нужно указать внутри скобок один аргумент — ключ, значение которого мы хотим получить. По своему функционалу метод схож с обращением к словарю по его ключу: словарь[ключ]. Однако, обращаясь к несуществующему ключу в словаре, мы получали ошибку KeyError. При использовании метода .get() такого происходить не будет. В этом случае метод вернет None, а программа продолжит свою работу.

В метод можно передать второй аргумент любого типа. Тогда при обращении к несуществующему в словаре ключу вместо None будет появляться значение этого аргумента. Давайте попробуем передать строку с сообщением  'Такого ключа нет' в качестве второго аргумента данного метода.

```python
d = {1: 'один', 2: 'два', 3: 'три'}

print(d.get(1))

print(d.get(1, 'Такого ключа нет'))

print(d.get(4))

print(d.get(4, 'Такого ключа нет'))
```

### Перебор элементов словаря

Важной особенностью словаря в Python является то, что он относится к коллекциям, так же как списки, строки и кортежи. А значит, по элементам этих коллекций можно итерироваться: перебирать все элементы при помощи цикла for. В этом нам помогут изученные методы словарей. По умолчанию в качестве объекта для перебора выступают ключи.

```python
capitals = {'Australia': 'Canberra',
            'Belgium': 'Brussels', 'Switzerland': 'Bern'}
for obj in capitals:
    print(obj)
```

Но что если мы хотим итерироваться не по ключам, а по значениям? Тогда это можно сделать двумя способами.

1. Уже знакомый нам метод .values()

```python
capitals = {'Australia': 'Canberra',

            'Belgium': 'Brussels', 'Switzerland': 'Bern'}

for obj in capitals.values():

    print(obj)
```

2. Итерироваться по ключам и обращаться к значениям.

```python
capitals = {'Australia': 'Canberra',
            'Belgium': 'Brussels', 'Switzerland': 'Bern'}
for key in capitals:
    print(capitals[key])
```

Python также позволяет итерироваться одновременно по ключам и значениям. Для этого нам понадобится метод .items().

```python
capitals = {'Australia': 'Canberra',

            'Belgium': 'Brussels', 'Switzerland': 'Bern'}

for obj in capitals.items():

    print(obj)
```

В данном случае переменная obj предоставляет кортеж, к элементам которого можно обратиться по индексам. Либо можно сразу распаковать пару ключ–значение на отдельные итераторы.

```python
capitals = {'Australia': 'Canberra',

            'Belgium': 'Brussels', 'Switzerland': 'Bern'}

for key, value in capitals.items():

    print(key, value)
```

# Типичные примеры использования словарей

##Подсчет количества элементов.
Словари можно использовать для подсчета количества элементов. Давайте попробуем посчитать, сколько раз встречается каждая буква в строке.

```python
my_string = input('Введите строку:')
my_dict = {}
for letter in my_string:
    if letter in my_dict:
        my_dict[letter] += 1
    else:
        my_dict[letter] = 1
print(my_dict)
for key, value in my_dict.items():
    print(key, value)
```

Получился неплохой алгоритм подсчета букв в строке. Однако у него есть несколько недостатков. Во-первых, знаки препинания, числа, а также все остальные символы он считает за буквы и добавляет в словарь. Во-вторых, алгоритм считает, что одна и та же буква в разных регистрах — это два разных символа.

Для лучшего решения этой задачи мы можем использовать следующие методы:


*   .isalpha() — метод строки, который позволит узнать, является ли символ буквой (вернет True) или нет (вернет False);
*   .lower() — метод строки, который переводит текст в нижний регистр;
*   .get() — изученный ранее метод словаря. Напомним, что он возвращает значение ключа. А если такого ключа в словаре нет, то вторым аргументом можно указать замену вывода None. Для нашего случая удобно вернуть 0.



Давайте рассмотрим код более детально.

```python
my_string = input('Введите строку:')
my_dict = {}
for letter in my_string:
    if letter.isalpha():  # Если символ является буквой
        low_letter = letter.lower()  # Переведем букву в нижний регистр
        # Проверяем, есть ли такая буква в словаре. Если да, то вернем значение и прибавим 1. Если нет, то вернем 0 и прибавим 1
        my_dict[low_letter] = my_dict.get(low_letter, 0) + 1
print(my_dict)
for key, value in my_dict.items():
    print(key, value)
```

Отлично! Теперь работает как надо!

##Установление соответствий
Словари, конечно же, можно использовать для установления соответствий между словами, как самый обычный словарь. Давайте составим небольшой англо-русский словарь и попробуем получить несколько значений из него!

```python
eng_rus_dict = {'car': 'автомобиль', 'school': 'школа',
                'to learn': ['учить', 'изучать']}
word = input('Введите слово для перевода на русский язык: ')
print(eng_rus_dict.get(word, 'Такого слова еще нет в словаре!'))
```

# Контейнер set (множество)

**Set** (множество) - контейнер, позволяющий хранить набор уникальных элементов. Эквивалентен математическому множеству.



Основные операции:
- добавить элемент в множество
- удалить элемент из множества
- проверить наличие элемента в множестве

Принцип работы также основан на хеш-таблице.

Где удобно использовать множества:
- подсчет уникальных элементов в списке
- быстрая проверка элемента на вхождение

**Замечание** - т.к. множество работает на хеш-таблицах, то в качестве элементов нельзя брать нехешируемые элементы!

### Создание множества

Создать множество в Python можно либо с помощью фигурных скобок (да, опять), либо с помощью функции `set`:

Создадим пустое множество:

```python
empty_set = set()
```

**Внимание!** Выражение вида `variable = {}` создает словарь:

```python
variable = {}
type(variable)
```

Создадим множество с элементами:

```python
set_example = {1, 2, 3}
set_example
```

```python
set_example = set([2, 3, 4, 3, 5, 2])
set_example
```

### Операции с множествами

В отличие от списков, словарей, кортежей и строк, получить прямой доступ к элементам множества мы не можем.

Для нас доступны лишь следующие операции:

```python
set_example = {1, '2', 3.0, 'four'}
set_example
```

1. добавить элемент в множество - метод `add`:

```python
set_example.add(5)
set_example
```

Множество хранит лишь уникальные элементы (нет дубликатов), если добавим уже существующий элемент, то множество не изменится:

```python
set_example.add(5)
set_example
```

2. удалить элемент из множества - для это есть 2 метода:

- метод `remove` - при попытке удалить уже отсутствующий в множестве элемент выдает ошибку:

```python
set_example.remove(1)
set_example
```

```python
set_example.remove(1)
set_example
```

- метод `discard` - при попытке удалить уже отсутствующий в множестве элемент **НЕ** выдает ошибку:

```python
set_example.discard(1)
set_example
```

3. проверить наличие элемента в множества - оператор `in` / `not in`:

```python
'2' in set_example
```

```python
1 in set_example
```

```python
1 not in set_example
```

### Математические операции над множествами

Множества в Python во многом "подражают" математическим множествам, поэтому для них определены разные операции над множествами:

```python
a = {1, 2, 3}
b = set([2, 3, 4])
```

- объединение двух множеств - метод `union`:sets-01.webp


```python
b.union(a)
```

- пересечение двух множеств - метод `intersection`:

image.png

```python
b.intersection(a)
```

- разность двух множеств (элементы множества `b`, которых нет в `a`) - метод `difference`:

image.png

```python
b.difference(a)
```

- симметричная разность (элементы двух множеств, которые встречаются лишь в одном из них) - метод `symmetric_difference`:

image.png

```python
b.symmetric_difference(a)
```

Эти же действия можно выполнять с помощью операций:

```python
a | b  # объединение
```

```python
a & b  # пересечение
```

```python
b - a  # разность
```

```python
b ^ a  # симметричная разность
```

Эти же методы и операторы имеют вариант с присвоением, например, найти пересечение `a` и `b` и записать в `a` (аналогично оператору `+=`):

- объединение - `|=`, `update()`
- пересечение - `&=`, `intersection_update()`
- разность -  `-=`, `difference_update()`
- симметричная разность - `^=`, `symmetric_difference_update()`

Помимо проверки на вхождение элемента в множество (`in`), имеются проверки на вхождения одного множества в другое:

```python
a = {1, 2, 3}
b = {1, 2, 3, 4}
```

```python
b.issubset(a)    # является ли b подмножеством a
```

```python
b.issuperset(a)  # является ли b надмножеством a
```

```python
b.isdisjoint(a)  # являются ли непересекающимися
```

### `frozenset`

Поскольку множество - изменяемый тип данных, его нельзя использовать в качестве ключа. Однако, у него есть свой аналог кортежа - **frozenset**, отличающееся лишь тем, что является неизменяемым:

```python
set_example = {1, 2, 3}
frozenset_example = frozenset(set_example)
```

```python
new_set = set(frozenset_example)
new_set.add(4)
```

```python
new_set
```

```python
set_example
```

```python
frozenset_example
```

Попробуем использовать их как ключи словаря:

```python
some_dict = {}
```

```python
some_dict[frozenset_example] = True
some_dict[frozenset_example]
```

```python
some_dict[set_example] = False
some_dict[set_example]
```
